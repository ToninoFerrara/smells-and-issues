```python
import csv
import os
import shutil
import sys
import subprocess
import time
import git
import pandas as pd
import numpy as np
from pydriller import RepositoryMining
from pydriller import GitRepository
import matplotlib.pyplot as plt

from github import Github
```

```python
dataset = pd.read_csv('data/smells.csv', delimiter=',', index_col=0)
print(dataset.shape)
```

```python
dataset.head()
```

```python
g = Github('matheusflauzino','xxxxxxx')

for index, row in dataset.iterrows():
    path_result = 'issues/' + row['path'].split('/')[1] + '.csv'
    print('repository {} issues...'.format(row['name']))
    if not (os.path.exists(path_result)):
        repository = g.get_repo(row['name'])
        name = row['name']
        #count_stars = repository.stargazers_count
        data_result = [] #clear variable data

        try:
            issues = repository.get_issues(state='all') #all issues
            for issue in issues: #interate issues
                data_result.append([name,getattr(issue,'state',''),getattr(issue,'title',''),list(getattr(issue,'labels','')),getattr(issue,'body','')])

            if not (os.path.exists(path_result)):
                with open(path_result, 'w') as csvfile:
                    spamwriter = csv.writer(csvfile, delimiter=',', quoting=csv.QUOTE_MINIMAL)
                    spamwriter.writerow(['repository','state','title','label','body'])
                    spamwriter.writerows(data_result)
        except:
            print('no exist issues')
```
